# Спорт Мастер

### Парсинг категории "Ракетки"

```python
import undetected_chromedriver as uc
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python


# Список для хранения данных
data = []

try:
    # Запуск браузера через undetected_chromedriver
    driver = uc.Chrome()

    # Перебор всех страниц
    for page in range(1, 5):  # Страницы от 1 до 4
        url = f"https://www.sportmaster.ru/catalog/vidy_sporta_/tennis/raketki/?watched=0&page={page}"
        driver.get(url)
        time.sleep(10)  # Ожидание загрузки страницы

        # Получение HTML страницы
        html = driver.page_source

        # Парсинг с помощью BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')

        # Находим контейнер с товарами
        products_container = soup.find('div', class_='sm-product-list__content__wrapper')
        products = products_container.find_all('li') if products_container else []

        # Извлекаем данные о каждом товаре
        for product in products:
            try:
                # Название товара
                name_tag = product.find('a', class_='sm-link sm-link_black')
                name = name_tag.text.strip() if name_tag else "Название отсутствует"

                # Цена товара
                price_tag = product.find('span', class_='sm-amount__value')
                price = price_tag.text.strip().replace('\xa0', '').replace('₽', '') if price_tag else "Цена отсутствует"

                # Добавляем в список
                data.append({'Название': name, 'Цена': price})
            except Exception as e:
                print(f"Ошибка обработки товара: {e}")
                continue

except Exception as ex:
    print(f"Ошибка: {ex}")

finally:
    driver.close()
    driver.quit()

# Преобразование списка в DataFrame
df = pd.DataFrame(data)

# Вывод результата
print(df)

# Сохраняем файл
df.to_csv('sportmaster_tennis_rackets.csv', index=True, encoding='utf-8')
```

### Запись в один файл

```python
# Словарь с названиями файлов и категориями
files = {
    "sportmaster_tennis_rackets.csv": "Ракетки",
    "sportmaster_tennis_myachi.csv": "Мячи",
    "sportmaster_tennis_ryukzaki_i_sumki.csv": "Рюкзаки",
    "sportmaster_tennis_struny.csv": "Струны",
    "sportmaster_tennis_namotka.csv": "Намотка",
    "sportmaster_tennis_vibrogasiteli.csv": "Виброгасители",
    "sportmaster_tennis_odezhda.csv": "Одежда",
    "sportmaster_tennis_odezhda2.csv": "Одежда_2",
    "sportmaster_tennis_obuv.csv": "Обувь",
    "sportmaster_tennis_inventar.csv": "Инвентарь"
}

# Список для объединения всех данных
combined_data = []

# Чтение и обработка каждого файла
for file_name, category in files.items():
    # Чтение файла
    df = pd.read_csv(file_name)
    
    # Добавление столбца "Категория"
    df['Категория'] = category
    
    # Добавление в общий список
    combined_data.append(df)

# Объединение всех данных в один DataFrame
sportmaster_df = pd.concat(combined_data, ignore_index=True)

# Сохранение в общий CSV файл
sportmaster_df.to_csv("sportmaster_data.csv", index=False, encoding="utf-8")

# Вывод первых строк результата
print(sportmaster_df.head())
```

# Saletennis

```python

# Список URL и соответствующих категорий
urls_and_categories = {
    "https://saletennis.com/catalog/tennisnye-raketki/": "Ракетки",
    "https://saletennis.com/catalog/tennisnye-myachi/": "Мячи",
    "https://saletennis.com/catalog/sumki-i-chekhly/": "Рюкзаки",
    "https://saletennis.com/catalog/tennisnye-struny/": "Струны",
    "https://saletennis.com/catalog/aksessuary/?params6%5B%5D=%D0%93%D1%80%D0%B8%D0%BF%D1%8B": "Намотка",
    "https://saletennis.com/catalog/aksessuary/?params6%5B%5D=%D0%92%D0%B8%D0%B1%D1%80%D0%BE%D0%B3%D0%B0%D1%81%D0%B8%D1%82%D0%B5%D0%BB%D0%B8": "Виброгасители",
    "https://saletennis.com/catalog/odezhda/": "Одежда_2",
    "https://saletennis.com/catalog/obuv/": "Обувь",
    "https://saletennis.com/catalog/stroitelstvo-kortov/": "Инвентарь",
}

# Список для хранения данных
data = []

try:
    # Запуск браузера через undetected_chromedriver
    driver = uc.Chrome()

    for url, category in urls_and_categories.items():
        try:
            driver.get(url)
            time.sleep(10)  # Ожидание загрузки страницы

            # Получение HTML страницы
            html = driver.page_source

            # Парсинг с помощью BeautifulSoup
            soup = BeautifulSoup(html, 'html.parser')

            # Находим контейнер с товарами
            products_container = soup.find('div', class_='content__cards')
            products = products_container.find_all('div', class_='c-item__wrapper c-item__wrapper--fixed') if products_container else []

            # Извлекаем данные о каждом товаре
            for product in products:
                try:
                    # Название товара
                    name_tag = product.find('a', class_='c-item__title')
                    name = name_tag.text.strip() if name_tag else "Название отсутствует"

                    # Цена товара
                    price_block = product.find('p', class_='c-item__price')
                    if price_block:
                        # Если есть старая цена, то новая находится после нее
                        old_price_tag = price_block.find('span', class_='c-item__price-old')
                        if old_price_tag:
                            # Новая цена - текст после <span>
                            new_price = price_block.text.replace(old_price_tag.text, '').strip().replace('\xa0', '').replace('₽', '')
                        else:
                            # Единственная цена
                            new_price = price_block.text.strip().replace('\xa0', '').replace('₽', '')
                    else:
                        new_price = "Цена отсутствует"

                    # Добавляем данные в список
                    data.append({'Название': name, 'Цена': new_price, 'Категория': category})
                except Exception as e:
                    print(f"Ошибка обработки товара: {e}")
                    continue

        except Exception as e:
            print(f"Ошибка обработки URL {url}: {e}")

except Exception as ex:
    print(f"Ошибка: {ex}")

finally:
    driver.close()
    driver.quit()

# Преобразование списка в DataFrame
df = pd.DataFrame(data)

# Вывод результата
print(df)

# Сохранение данных в CSV
df.to_csv('saletennis_data.csv', index=False, encoding='utf-8')
```

```python
# Список для хранения данных
data = []

try:
    # Запуск браузера через undetected_chromedriver
    driver = uc.Chrome()

    # Перебор всех страниц
    for page in range(1, 6):  # Страницы от 1 до 5
        url = f"https://tennisdirect.ru/raketki/page-{page}/"
        driver.get(url)
        time.sleep(7)  # Ожидание загрузки страницы

        # Получение HTML страницы
        html = driver.page_source

        # Парсинг с помощью BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')

        # Находим контейнер с товарами
        products_container = soup.find('div', class_='grid-list')
        products = products_container.find_all('div', class_='ty-column4') if products_container else []

        # Извлекаем данные о каждом товаре
        for product in products:
            try:
                # Название товара
                name_tag = product.find('a', class_='product-title')
                name = name_tag.get('title', "Название отсутствует") if name_tag else "Название отсутствует"

                # Цена товара (исключаем копейки)
                price_tag = product.find('span', class_='ty-price-num')
                if price_tag:
                    # Извлекаем текст до тега <sup>
                    price = price_tag.contents[0].strip().replace('\xa0', '').replace('₽', '')
                else:
                    price = "Цена отсутствует"

                # Добавляем в список
                data.append({'Название': name, 'Цена': price})
            except Exception as e:
                print(f"Ошибка обработки товара: {e}")
                continue


except Exception as ex:
    print(f"Ошибка: {ex}")

finally:
    driver.close()
    driver.quit()

# Преобразование списка в DataFrame
df = pd.DataFrame(data)

# Вывод результата
print(df)

# Сохранение в CSV-файл
df.to_csv('tennisdirect_raketki.csv', index=True, encoding='utf-8')
```

```python
import pandas as pd
import re
from difflib import SequenceMatcher
```

```python
# Функция для очистки названий товаров
def clean_product_name(name):
    # Оставляем только латинские буквы, цифры и пробелы, приводим к нижнему регистру
    return re.sub(r'[^a-zA-Z0-9\s]', '', str(name)).lower()

# Функция для расчёта текстового сходства
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

# Загрузка данных
tennis_store_data = pd.read_csv("tennis_store_data.csv")
saletennis_data = pd.read_csv("saletennis_data.csv")
sportmaster_data = pd.read_csv("sportmaster_data.csv")
tennisdirect_data = pd.read_csv("tennisdirect_data.csv")
tennispro_data = pd.read_csv("tennis-pro_data.csv")

# Очистка названий товаров
for df in [tennis_store_data, saletennis_data, sportmaster_data, tennisdirect_data, tennispro_data]:
    df['cleaned_name'] = df['Название'].apply(clean_product_name)

# Уникальные очищенные названия товаров из tennis_store_data
tennis_store_names = tennis_store_data['cleaned_name'].unique()


```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Функция для очистки названий товаров
def clean_product_name(name):
    # Оставляем только латинские буквы, цифры и пробелы, приводим к нижнему регистру
    return re.sub(r'[^a-zA-Z0-9\s]', '', str(name)).lower()

# Загрузка данных
tennis_store_data = pd.read_csv("tennis_store_data.csv")
saletennis_data = pd.read_csv("saletennis_data.csv")
sportmaster_data = pd.read_csv("sportmaster_data.csv")
tennisdirect_data = pd.read_csv("tennisdirect_data.csv")
tennispro_data = pd.read_csv("tennis-pro_data.csv")

# Очистка названий товаров
for df in [tennis_store_data, saletennis_data, sportmaster_data, tennisdirect_data, tennispro_data]:
    df['cleaned_name'] = df['Название'].apply(clean_product_name)

# Уникальные очищенные названия товаров из tennis_store_data
tennis_store_names = tennis_store_data['cleaned_name'].unique()

# Функция для расчёта сходства по косинусной метрике для каждого товара из tennis_store с товаром из другого магазина
def count_similar_names_with_tfidf(store_names, compare_names, threshold=0.8):
    count = 0
    # Векторизация для store_names и compare_names (для каждого магазина отдельно)
    vectorizer = TfidfVectorizer()
    store_vectors = vectorizer.fit_transform(store_names)  # Векторизация store_names
    compare_vectors = vectorizer.transform(compare_names)  # Векторизация названий товаров другого магазина
    
    # Расчёт сходства между каждым товаром из store_names и каждым товаром из compare_names
    similarities = cosine_similarity(store_vectors, compare_vectors)
    
    for i in range(similarities.shape[0]):  # Для каждого товара из tennis_store_names
        if any(similarity >= threshold for similarity in similarities[i]):  # Если есть хотя бы одно совпадение
            count += 1
    return count

# Подсчёт пересечений для каждого магазина
intersections = {
    "saletennis": count_similar_names_with_tfidf(tennis_store_names, saletennis_data['cleaned_name']),
    "sportmaster": count_similar_names_with_tfidf(tennis_store_names, sportmaster_data['cleaned_name']),
    "tennisdirect": count_similar_names_with_tfidf(tennis_store_names, tennisdirect_data['cleaned_name']),
    "tennispro": count_similar_names_with_tfidf(tennis_store_names, tennispro_data['cleaned_name']),
}

# Вывод результатов
print("Количество пересечений товаров (с использованием TF-IDF сходства):")
for store, count in intersections.items():
    print(f"{store}: {count}")
```
